```python
df=pd.read_csv('boston.csv')
```

```python
df.head()
```

```python
print(df.shape)
```

```python
df.columns
```

```python
df.groupby('chas')['medv'].mean()# controllando prezzo per distanza fiume
```

```python
df['crim'].corr(df['medv'])#correlazione criminalita/prezzo
```

```python
# ce una correlazione negativa moderata, quando la criminalita aumenta , il prezzo diminuisce,
# ossia, le case custano meno dove la criminalita è piu alta.
```

```python
df.groupby('ptratio')['rm'].mean() # media do numero de stanze rispetto al rapporto alunni/insegnante
```

```python
df.groupby('ptratio')['medv'].mean()# valore medio case rispetto al rapporto alunni insegnante
```

```python
df['ptratio'].corr(df['rm']) #correlazione tra numero di alunni per insegnante e stanze
                        #correlazione negativa moderata,numero alunni/insegnante aumenta il numero di estanze 
                        #della casa diminuisce, ossia dove ci sono scuole peggiore si tende a havere case piu 
                        #piccole
```

```python
df['ptratio'].corr(df['medv']) #correlazione tra numero di alunni per insegnante e prezzo delle casse
                            #correlazione negativa forte, nuero alunni/insegnante aumenta o prezzo delle
                            # case diminuisce in modo significativo, ossia dove ci sono scuole peggiore, 
                            # le case valgono meno.
```

```python
df.groupby('rad')['age'].mean()#media di eta delle casa cambia rispetto acessibilita alle autostrade
```

```python
df['rad'].corr(df['age'])#correlazione positiva tra acessibilita autostrade e idade delle case
                        # quanto maggiore il RAD  maggiore è eta delle casa
```

```python
df.groupby('rad')['rm'].mean()# rad e numero di stanze
```

```python
df.groupby('rad')['tax'].mean()# piu aumento il rad piu aumentano le tasse
```
